```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the network architecture
def neural_network(x):
    w1, b1 = 1000, -400
    w2, b2 = 1000, -600
    h1 = sigmoid(w1 * x + b1)
    h2 = sigmoid(w2 * x + b2)
    w_h1,w_h2 = 1,-1
    output = w_h1*h1 + w_h2* h2
    return output

# Generate x values
x = np.linspace(0, 1, 1000)

# Calculate network outputs
y = neural_network(x)

# Plot the approximation
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Bump Function Approximation')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the network architecture
def neural_network(x1, x2):
    w1_x1, b1_x1 = 1000, -300
    w2_x1, b2_x1 = 1000, -700
    h1 = sigmoid(w1_x1 * x1 + b1_x1)
    h2 = sigmoid(w2_x1 * x1 + b2_x1)
    w_h1,w_h2 = 1,-1
    output = w_h1*h1 + w_h2* h2
    return output

# Generate x1 and x2 values
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100)
X1, X2 = np.meshgrid(x1, x2)

# Calculate network outputs
Z = neural_network(X1, X2)

# Plot the 2D bump function approximation
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x)')
ax.set_title('2D Bump Function Approximation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the network architecture
def neural_network(x1, x2):
    # First Hidden Layer
    w1_x1, w1_x2, b1 = 1000, 0, -450
    w2_x1, w2_x2, b2 = 1000, 0, -550  
    w3_x1, w3_x2, b3 = 0, 1000, -450  
    w4_x1, w4_x2, b4 = 0, 1000, -550
    h1 = sigmoid(w1_x1 * x1 + w1_x2 * x2 + b1)
    h2 = sigmoid(w2_x1 * x1 + w2_x2 * x2 + b2)
    h3 = sigmoid(w3_x1 * x1 + w3_x2 * x2 + b3)
    h4 = sigmoid(w4_x1 * x1 + w4_x2 * x2 + b4)
    
    # Second Hidden Layer
    w_h1,w_h2,w_h3,w_h4 = 1000,-1000,1000,-1000 
    b5 = -1500
    h5 = sigmoid(w_h1*h1 + w_h2*h2 + w_h3*h3 + w_h4*h4 + b5)
    return h5



# Generate x1 and x2 values
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100)
X1, X2 = np.meshgrid(x1, x2)

# Calculate network outputs
Z = neural_network(X1, X2)

# Plot the 2D tower function approximation
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x)')
ax.set_title('2D Tower Function Approximation')
plt.show()
```
